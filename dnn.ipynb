```python
# Importing libraries
import pandas as pd    
import numpy as np
# Reading data to Dataframes
    
data_train = pd.read_csv('../input/health-insurance-cross-sell-prediction/train.csv')
data_test = pd.read_csv('../input/health-insurance-cross-sell-prediction/test.csv')
```

```python
data_train = pd.get_dummies(data_train)
data_test = pd.get_dummies(data_test)
```

```python
data_train.head()
```

```python
X_train = data_train.iloc[:,1:8]   
y_train = data_train.iloc[:,8]     
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(data_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
```

```python
#Input layer with 5 inputs neurons
classifier.add(Dense(3,  activation = 'relu', input_dim = 7))
#Hidden layer
classifier.add(Dense(5,  activation = 'relu'))
classifier.add(Dense(5,  activation = 'relu'))
#output layer with 1 output neuron which will predict 1 or 0
classifier.add(Dense(1,  activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 10)
```
